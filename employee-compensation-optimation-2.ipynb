## 1. Define all range

```python
# Performance Rating
performance_range = {
    1: (0.00, 0.70),
    2: (0.70, 1.00),
    3: (1.00, 1.30),
    4: (1.30, 1.50)
}
```

```python
# JobInvolvement
jobinv_range = {
    1: (0.85, 0.95),
    2: (0.95, 1.05),
    3: (1.05, 1.15),
    4: (1.15, 1.25)
}
```

```python
# Department
dept_range = {
    "Sales": (1.15, 1.25),
    "Research & Development": (1.05, 1.15),
    "Human Resources": (0.95, 1.05)
}
```

```python
# JobLevel
joblevel_range = {
    1: (0.85, 1.10),
    2: (0.90, 1.15),
    3: (0.95, 1.20),
    4: (1.00, 1.25),
    5: (1.05, 1.30)
}
```

```python
# YearsAtCompany
def get_tenure_multiplier(years):
    if years < 1:
        return 0.5
    elif years < 2:
        return 1.0
    elif years < 4:
        return 1.1
    elif years < 6:
        return 1.2
    elif years < 8:
        return 1.3
    elif years < 10:
        return 1.4
    else:
        return 1.5
```

### 1.2 Range Department one-hot encoding

```python
def get_department_multiplier(row, score):
    if row.get("Department_Sales", 0) == 1:
        low, high = dept_range["Sales"]
    elif row.get("Department_Research & Development", 0) == 1:
        low, high = dept_range["Research & Development"]
    else:
        low, high = dept_range["Human Resources"]

    return map_to_range(low, high, score)
```

```python
import numpy as np
def generate_target_bonus(row):
    if row["PerformanceRating"] == 4:
        low, high = 140, 150
    elif row["PerformanceRating"] == 3:
        low, high = 100, 120
    else:
        low, high = 0, 70

    seed = row["JobInvolvement"] + (2 if row["OverTime"] == 1 else 0)
    variation = (seed / 6) * (high - low)

    return low + variation
    
df["BonusScore"] = df.apply(generate_target_bonus, axis=1)
df["BonusScore"] = df["BonusScore"] + np.random.normal(0, 5, len(df))
```

## 3. Convert score multiplier

```python
def map_to_range(low, high, score):
    return low + (high - low) * score
```

## 4. Compute all multiplier

```python
joblevel_bonus_range = {
    1: (1.0, 1.5),
    2: (1.0, 1.5),
    3: (1.5, 2.0),
    4: (2.0, 2.5),
    5: (2.5, 3.0)
}
```

```python
def compute_bonus_rate(row, score):
    performance = map_to_range(*performance_range[row["PerformanceRating"]], score)
    jobinv = map_to_range(*jobinv_range[row["JobInvolvement"]], score)
    dept = get_department_multiplier(row, score)
    joblevel = map_to_range(*joblevel_range[row["JobLevel"]], score)
    tenure = get_tenure_multiplier(row["YearsAtCompany"])

    # tetap pakai multiplicative
    raw_bonus_rate = (
        performance *
        jobinv *
        dept *
        (joblevel ** 0.1) *
        tenure
    )

    # normalize ke 0–1
    normalized = min(raw_bonus_rate / 3.0, 1.0)

    # mapping ke range JobLevel
    low, high = joblevel_bonus_range[row["JobLevel"]]
    bonus_rate = low + (high - low) * normalized

    return bonus_rate
```

```python
X_test = X_test.copy()
X_test["score"] = model_scores

final_bonus = []

for _, row in X_test.iterrows():
    bonus_rate = compute_bonus_rate(row, row["score"])
    final_bonus.append(row["MonthlyIncome"] * bonus_rate)

X_test["FinalBonus"] = final_bonus
```
